# 네이버 부동산 매물 수집 가이드

## 📌 대상 사이트
- 네이버 부동산: [https://land.naver.com/](https://land.naver.com/)

---

## 📌 수집 대상 설정
1. 네이버 부동산 접속
2. '매물' 탭으로 이동
3. 원하는 지역 선택 (예: 한남동 한남더힐)
4. '전체거래방식'을 **매매**로 설정
5. '동일매물 묶기' 체크 활성화

---

## 📌 데이터 요청 분석 방법
1. **F12**를 눌러 개발자 도구 실행
2. **Network 탭** 선택
3. **Fetch/XHR** 필터 선택
4. 네이버 부동산 페이지에서 매물 검색
5. 요청 리스트에서 `realEstateType`이 포함된 요청 확인
6. 해당 요청에서 우클릭 → **Copy** → **Copy as cURL (bash)** 선택

---

## 📌 cURL 변환
1. 아래 사이트 접속: [https://curlconverter.com/](https://curlconverter.com/)

````python
from PySide6.QtWidgets import QApplication, QMainWindow, QWidget, QVBoxLayout, QHBoxLayout, QPushButton, QLabel, QSpinBox, QTableWidget, QTableWidgetItem, QMessageBox
from PySide6.QtCore import Qt
import sys
import requests
import pandas as pd
import json

class PropertyCrawlerGUI(QMainWindow):
    """
    네이버 부동산 매물 데이터를 수집하는 GUI 애플리케이션
    """
    def __init__(self):
        super().__init__()
        self.setWindowTitle("네이버 부동산 매물 수집기")
        self.setMinimumSize(800, 600)

        # 메인 위젯과 레이아웃 설정
        main_widget = QWidget()
        self.setCentralWidget(main_widget)
        layout = QVBoxLayout(main_widget)

        # 컨트롤 영역 생성
        control_layout = QHBoxLayout()
        
        # 페이지 수 입력
        self.page_label = QLabel("수집할 페이지 수:")
        self.page_spinner = QSpinBox()
        self.page_spinner.setRange(1, 100)
        self.page_spinner.setValue(1)
        
        # 수집 버튼
        self.collect_button = QPushButton("데이터 수집")
        self.collect_button.clicked.connect(self.collect_data)
        
        # 저장 버튼
        self.save_button = QPushButton("엑셀로 저장")
        self.save_button.clicked.connect(self.save_to_excel)
        self.save_button.setEnabled(False)

        # 컨트롤 위젯들을 레이아웃에 추가
        control_layout.addWidget(self.page_label)
        control_layout.addWidget(self.page_spinner)
        control_layout.addWidget(self.collect_button)
        control_layout.addWidget(self.save_button)
        control_layout.addStretch()

        # 테이블 위젯 설정
        self.table = QTableWidget()
        self.table.setColumnCount(16)
        headers = ['매물번호', '건물명', '거래유형', '층수정보', '가격', '면적(㎡)', 
                  '전용면적(㎡)', '방향', '등록일', '특징', '태그', '건물동', 
                  '중개사무소', '위도', '경도', '협회']
        self.table.setHorizontalHeaderLabels(headers)

        # 메인 레이아웃에 위젯 추가
        layout.addLayout(control_layout)
        layout.addWidget(self.table)

        # 데이터 저장용 변수
        self.property_data = None

        # API 설정
        self.cookies = {
            'nhn.realestate.article.rlet_type_cd': 'A01',
            'nhn.realestate.article.trade_type_cd': '""',
            'nhn.realestate.article.ipaddress_city': '1100000000',
            '_fwb': '67WsNzBLmId2QXGl07hvtO.1740895958329',
            'NAC': 's4JFBgwkqwjC',
            'landHomeFlashUseYn': 'Y',
            'NACT': '1',
            'NNB': 'AJ3YDBWV63BWO',
            'SRT30': '1740903507',
            'SRT5': '1740904040',
            'REALESTATE': 'Sun%20Mar%2002%202025%2017%3A29%3A27%20GMT%2B0900%20(Korean%20Standard%20Time)',
            'BUC': 'CZ8tyr7lKRyotLTGCVFGkcHmWAWyniTVHxLEcjED6bs=',
        }

        self.headers = {
            'accept': '*/*',
            'accept-language': 'ko-KR,ko;q=0.9',
            'authorization': 'Bearer eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJpZCI6IlJFQUxFU1RBVEUiLCJpYXQiOjE3NDA5MDQxNjcsImV4cCI6MTc0MDkxNDk2N30.jUMvmEC53S6Y0Y5u1S3LfhJMmd5-rqYqQpGczNhtobk',
            'priority': 'u=1, i',
            'referer': 'https://new.land.naver.com/complexes/102737?ms=37.5373135,127.0098913,17&a=APT:ABYG:JGC:PRE&e=RETAIL&ad=true',
            'sec-ch-ua': '"Not(A:Brand";v="99", "Google Chrome";v="133", "Chromium";v="133"',
            'sec-ch-ua-mobile': '?0',
            'sec-ch-ua-platform': '"Windows"',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36',
        }

    def collect_data(self):
        """데이터 수집 실행"""
        try:
            max_pages = self.page_spinner.value()
            self.property_data = self.get_property_data(max_pages)
            self.update_table()
            self.save_button.setEnabled(True)
            QMessageBox.information(self, "성공", f"{len(self.property_data)}개의 매물 데이터를 수집했습니다.")
        except Exception as e:
            QMessageBox.critical(self, "오류", f"데이터 수집 중 오류가 발생했습니다: {str(e)}")

    def get_property_data(self, max_pages):
        """매물 데이터 수집"""
        all_property_data = []
        
        for page in range(1, max_pages + 1):
            try:
                response = requests.get(
                    f'https://new.land.naver.com/api/articles/complex/16378?realEstateType=APT%3AABYG%3AJGC%3APRE&tradeType=&tag=%3A%3A%3A%3A%3A%3A%3A%3A&rentPriceMin=0&rentPriceMax=900000000&priceMin=0&priceMax=900000000&areaMin=0&areaMax=900000000&oldBuildYears&recentlyBuildYears&minHouseHoldCount&maxHouseHoldCount&showArticle=false&sameAddressGroup=true&minMaintenanceCost&maxMaintenanceCost&priceType=RETAIL&directions=&page={page}&complexNo=16378&buildingNos=&areaNos=&type=list&order=rank',
                    cookies=self.cookies,
                    headers=self.headers,
                )
                
                data = response.json()
                articles = data['articleList']
                
                if not articles:
                    break
                    
                for article in articles:
                    property_info = {
                        '매물번호': article['articleNo'],
                        '건물명': article['articleName'],
                        '거래유형': article['tradeTypeName'],
                        '층수정보': article['floorInfo'],
                        '가격': article['dealOrWarrantPrc'],
                        '면적(㎡)': article['area1'],
                        '전용면적(㎡)': article['area2'],
                        '방향': article['direction'],
                        '등록일': article['articleConfirmYmd'],
                        '특징': article['articleFeatureDesc'],
                        '태그': ', '.join(article['tagList']),
                        '건물동': article['buildingName'],
                        '중개사무소': article['realtorName'],
                        '위도': article['latitude'],
                        '경도': article['longitude'],
                        '협회': article['cpName']
                    }
                    all_property_data.append(property_info)
                
            except Exception as e:
                raise Exception(f"페이지 {page} 데이터 수집 중 오류: {str(e)}")
                
        return all_property_data

    def update_table(self):
        """테이블 위젯 업데이트"""
        if not self.property_data:
            return
            
        self.table.setRowCount(len(self.property_data))
        
        for row, data in enumerate(self.property_data):
            for col, (key, value) in enumerate(data.items()):
                item = QTableWidgetItem(str(value))
                self.table.setItem(row, col, item)
        
        self.table.resizeColumnsToContents()

    def save_to_excel(self):
        """데이터를 엑셀 파일로 저장"""
        try:
            if self.property_data:
                df = pd.DataFrame(self.property_data)
                df.to_excel('naver_property_data.xlsx', index=False)
                QMessageBox.information(self, "성공", "데이터가 엑셀 파일로 저장되었습니다.")
        except Exception as e:
            QMessageBox.critical(self, "오류", f"파일 저장 중 오류가 발생했습니다: {str(e)}")

if __name__ == '__main__':
    app = QApplication(sys.argv)
    window = PropertyCrawlerGUI()
    window.show()
    sys.exit(app.exec())
````

## ✅ 부동산 매매 데이터 크롤링 프로그램의 장점

### 1️⃣ 시세 및 트렌드 분석 가능
- 실시간 데이터 수집을 통해, 특정 지역의 가격 변동 추이를 직접 파악할 수 있음.
- 매도/매수자 동향 파악 (매물 수 급증 → 매도세 증가, 매물 부족 → 매수세 증가)
- 단순 시세 비교가 아닌, 매물별 특성 (층수, 방향, 내부 옵션 등)까지 분석 가능.

### 2️⃣ 데이터 독점성 확보
- 직접 수집한 데이터는 경쟁자들이 접근할 수 없는 "나만의 데이터"가 됨.
- 네이버 부동산 등 외부 사이트는 데이터 제공 범위가 제한적이지만, 크롤링으로는 원하는 데이터 항목까지 모두 확보 가능.
- 향후 AI 분석, 투자 추천 서비스 등에 활용할 원천 데이터 확보.

### 3️⃣ 매매 타이밍 포착
- 가격 급등/급락 시 실시간 알림 기능 추가 가능.
- 관심 단지, 관심 지역에 대한 매물 변화 감지 → 투자 타이밍 캐치.
- 뉴스나 정책 변화와 실제 매물 가격 변화의 상관관계 분석도 가능.

### 4️⃣ 실거래가 데이터와 매물 데이터 비교 분석
- 정부에서 제공하는 실거래가 데이터와 네이버 부동산 매물 데이터를 함께 분석.
- "허위매물" 필터링 가능 (시세보다 지나치게 낮거나 높은 매물 탐지)
- 매물 가격과 실제 거래 가격의 괴리 분석으로, 협상 전략 수립 가능.

### 5️⃣ 자동화된 리포트 생성
- 일별, 주별, 월별로 지역별 시세 및 매물 추이 리포트 자동 생성.
- 투자 보고서, 상담 자료로 바로 활용 가능.
- 데이터 시각화까지 연결해서, 부동산 인사이트를 시각적으로 표현.

### 6️⃣ 빠른 의사결정 및 실시간 대응
- 사람이 직접 사이트에서 매물 확인하는 대신, 프로그램이 24시간 자동 모니터링.
- 급매물, 호재 지역 매물 즉시 포착 가능.
- 매물 변화, 가격 변동을 알림톡/메일로 실시간 전달.

### 7️⃣ 투자 시뮬레이션 및 분석 고도화
- 매물 데이터 + 대출 이자율 + 취득세/양도세 등 추가 정보 결합.
- 투자 수익률 계산, 보유 기간별 시나리오 분석.
- 투자 가이드라인 자동 추천 (매수/매도 추천)
