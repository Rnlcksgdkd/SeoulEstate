```python
import folium

# 모서리 4개 좌표 (위도, 경도)
corner_coords = [
    (37.4735804, 127.0380378),  # 좌상
    (37.4735804, 127.042544),  # 우상
    (37.4719498, 127.042544),  # 우하
    (37.4719498, 127.0380378)   # 좌하
]

# 중심 좌표 계산 (평균)
center_lat = sum([coord[0] for coord in corner_coords]) / 4
center_lon = sum([coord[1] for coord in corner_coords]) / 4

# 지도 생성
m = folium.Map(location=[center_lat, center_lon], zoom_start=15)

# 마커 표시
for coord in corner_coords:
    folium.Marker(location=coord).add_to(m)

# 사각형(Polygon) 그리기
folium.Polygon(
    locations=corner_coords,
    color='blue',        # 테두리 색상
    weight=2,            # 테두리 두께
    fill=True,
    fill_color='lightblue',
    fill_opacity=0.4
).add_to(m)

# 결과 저장 또는 표시
m
```
